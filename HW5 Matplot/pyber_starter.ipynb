```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_load = "data/city_data.csv"
ride_data_load = "data/ride_data.csv"

# Read the City and Ride Data

city_data = pd.read_csv(city_data_load)
ride_data = pd.read_csv(ride_data_load)
city_data.head(10)

# Combine the data into a single dataset

data_merge = pd.merge(city_data, ride_data, how="left", on=["city","city"])
data_merge.head()
```

```python
urban = data_merge.loc[data_merge["type"] == "Urban", :]
rural = data_merge.loc[data_merge["type"] == "Rural", :]
suburban = data_merge.loc[data_merge["type"] == "Suburban", :]
```

```python
group_urban = urban.groupby(['city']).count()
group_rural = rural.groupby(['city']).count()
group_suburban = suburban.groupby(['city']).count()
size = data_merge['driver_count']
```

## Total Fares by City Type

```python
# Calculate Type Percents

# Build Pie Chart

# Save Figure

urban_fare = urban['fare'].sum()
rural_fare = rural['fare'].sum()
suburban_fare = suburban['fare'].sum()

labels = ["Rural", "Urban", "Suburban"]

sizes = [rural_fare, urban_fare, suburban_fare]
explode = (0, 0.1, 0)

# The colors of each section of the pie chart
colors = ["yellow", "red", "lightskyblue"]

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=260)

plt.axis("equal")

#flatten
plt.axis("equal")
plt.title("% of Total Fares by City Type")
plt.show()
```

```python
# Show Figure
plt.show()
```
